```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 6 * np.pi, 1000)
f = [t * np.cos(t), t * np.sin(t)]

t = np.linspace(0, np.pi, 10)
f_lin = [-np.pi / 2 * (t - np.pi / 2),
         np.pi / 2 + (t - np.pi) / 2]

plt.figure(figsize=(6,6))
plt.plot(*f, 'r')
plt.plot(*f_lin, 'g')
plt.axis('equal')
plt.show()
```
