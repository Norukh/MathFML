```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-2, 3, 1000)

# a(t) = (t^2, t^4)
x = t ** 2
y = t ** 4
plt.plot(x, y, "r")

# b(t) = (t^3, t^4)
x = t ** 3
y = t ** 4
plt.plot(x, y, "g")

# c(t) = (t^2, t)
x = t ** 2
y = t
plt.plot(x, y, "b")

# d(t) = (e^t, ln(t))
# only positive domain
t = np.linspace(0.00000001, 3, 100)
x = np.exp(t)
y = np.log(t)
plt.plot(x, y, "m")

#plt.axis("equal")
plt.axis([-10, 10, -10, 10])

plt.axhline(0)
plt.axvline(0)
plt.show()
```
